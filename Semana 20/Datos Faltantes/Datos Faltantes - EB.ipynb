# Rellenar NAN

```python
#Contamos cuantos  nos faltan
pd.isnull(titanic["age"]).values.ravel().sum()
```

```python
# permite obtener la media
titanic["age"].mean()
```

hay 263 personas a la que se les desconoce su edad

```python
# Vamos a rrellenar la edad con los promedios de las edades que si conocemos
# fillna se usa para llenar datos faltantes

col = titanic["age"].fillna(titanic["age"].mean())
```
