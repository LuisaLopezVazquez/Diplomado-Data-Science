```python
#Convertimos a numericas las columnas 
df['red_blood_cell_count'] = pd.to_numeric(df['red_blood_cell_count'], errors='coerce')
df['packed_cell_volume'] = pd.to_numeric(df['packed_cell_volume'], errors='coerce')
df['white_blood_cell_count'] = pd.to_numeric(df['white_blood_cell_count'], errors='coerce')
```

```python
df.info()
```

```python
df.drop(['id'], axis=1, inplace=True)
```

```python
df.head()
```

```python
#Creamos el ciclo
numerical_features=[]  #Guardar los valores que son numericos
categorical_features=[] # Guardar los valores que son categoricos

for i in df.drop('classification', axis=1).columns:
    
    if df[i].nunique()>7:
        numerical_features.append(i)
        print(numerical_features,i)
    else:
        categorical_features.append(i)
        print(categorical_features,i)

```

```python
numerical_features
```

```python
#Replazar  valores incorrectos
df['diabetes_mellitus'] = df['diabetes_mellitus'].replace(to_replace = {'\tno':'no','\tyes':'yes',' yes':'yes'})
df['coronary_artery_disease'] = df['coronary_artery_disease'].replace(to_replace = '\tno', value='no')
df['classification'] = df['classification'].replace(to_replace = 'ckd\t', value = 'ckd')
```

```python
#Mostramos la suma de los datos nulos que tenemos en nuestra lista de caracteristicas categoricas
df.loc[:,categorical_features].isnull().sum().sort_values(ascending=False)
```

```python
#Mostramos la suma de los datos nulos que tenemos en nuestra lista de caracteristicas numericas
df.loc[:,numerical_features].isnull().sum().sort_values(ascending=False)
```

```python
#Eliminamos classification
df_v2=df.drop('classification', axis=1)
```

## Imputaci칩n de caracter칤sticas num칠ricas usando MICE

```python
# MICE requiere valores flotantes
# Convertimos las caracteristicas numericas a flotantes
df_entrenamiento_numericos =  df_entrenamiento.loc[:,numerical_features].astype('float64')
```

```python
df_entrenamiento_numericos.head(2)
```

```python
# Pasar las matrices numpy a los mice
df_entrenamiento_numericas_imputacion= mice(df_entrenamiento_numericos.values) # Le pasamo el data frame para la imputacion
```

```python
#Vemos cuantas valores nulos tenemos en nuestro dataFrame
```

```python
#Ahora a nuestro dataFrame original le vamos a pasar los valores que 
# se obtuvieron de hacer la inputacion con mice
df_entrenamiento.loc[:,numerical_features] = df_entrenamiento_numericas_imputacion
```

```python
#Imprimimos para ver el resultado final
df_entrenamiento.loc[:,numerical_features].isna().sum().sort_values(ascending=False)
```
