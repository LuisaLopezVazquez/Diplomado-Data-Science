```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

```python
#Creamos la conexion
conexion= sqlite3.connect('punto_venta.db')
cursor=conexion.cursor()
```

```python
#Consultamos la tabla de ventas
df_ventas = pd.read_sql_query('SELECT * FROM ventas', conexion)
df_ventas.head()
```

```python
#Consultamos la tabla de clientes
df_clientes = pd.read_sql_query('SELECT * FROM clientes', conexion)
df_clientes.head()
```

```python
df_clientes_y_compras = pd.read_sql_query("SELECT clientes.id_cliente,nombre,apellido_paterno,apellido_materno,monto,total_productos,forma_pago FROM clientes"
                                +" INNER JOIN ventas ON clientes.id_cliente=ventas.id_cliente;", conexion)
df_clientes_y_compras.head()
```

```python
# Merge une las tablas
pd.merge(df_ventas,df_clientes_y_compras, how='outer', left_index=True, right_index=True)
```

```python
df_ventas=df_ventas.set_index('id_cliente')
```

```python
# Se elimina el indice del data frame
pd.merge(df_ventas,df_clientes, how='left', left_index=True, right_index=True)
```

```python
#Merge permite unir dos data frame 
```

```python
#Merge permite unir dos data frame 
pd.merge(df_ventas,df_clientes, how='right', left_index=True, right_index=True)
```

```python
df_clientes=df_clientes.set_index('id_cliente')
```

```python
#Merge permite unir dos data frame 
pd.merge(df_ventas,df_clientes, how='inner', left_index=True, right_index=True)
```

```python
df_clientes=df_clientes.reset_index()
```

```python
pd_ventas_clientes = pd.merge(df_ventas,df_clientes, how='inner', on=['id_cliente'])
pd_ventas_clientes.head()
```

```python
#Cerrar su conexin a la base de datos!
conexion.close()
```
