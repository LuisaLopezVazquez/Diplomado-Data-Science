<h1 style="color:purple;">Datos Categóricos a Visualizar</h1> 

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="ticks", context="talk", palette="tab10")

consumos = sns.load_dataset("tips")

consumos.columns = ["total", "propina", "genero", "fumador",
                    "dia", "tipo", "comensales"]
consumos
```

```python
sns.catplot(x='dia',y ='total', data=consumos, kind='box', col='fumador')
plt.show()
```

<h1 style="color:blue;"> Gráfica boxenplot.. </h1>

```python
sns.catplot(x='dia', y='total', data=consumos, kind='boxen', col='fumador')
plt.show()
```

Estas graficas son especializadas es remendable cuando se tiene muchos datos   
Es conveniente usarlas con personas que ya conocen y saben interpretarlas

<h2 style="color:blue;"> Gráfica violinplot  </h2>

```python
sns.catplot(x='dia' ,y='total', data=consumos, kind='violin')
plt.show()
```

```python
sns.catplot(x='dia' ,y='total', data=consumos, kind='violin', hue='fumador')
plt.show()
```

```python
sns.catplot(x='dia' ,y='total', data=consumos, kind='violin', hue='fumador', split=True)
plt.show()
```

<h2 style="color:indianred;"> Gráfica barplot</h2>

```python
sns.catplot(x='dia' , y='total', data=consumos, kind='bar', hue='comensales')
plt.show()
```

<h2 style="color:indianred;"> Gráfica countplot </h2>

```python
sns.catplot(x='dia', data=consumos, kind='count', hue='comensales')
plt.show()
```
