```python
#Importamos las librerias
import lightgbm as lgb
from sklearn.datasets import load_iris
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from pandas import DataFrame
from numpy import argmax
```

```python
#Se instala la libreria en caso de no tenerla 
#pip install lightgbm 
```

### Preparando los datos

    Usamos el conjunto de datos de Iris como datos de clasificación objetivo y podemos cargarlo fácilmente desde el módulo sklearn.datasets. Para mantener los nombres de las columnas de características, usaré el tipo Pandas DataFrame para los datos de características.

```python
iris = load_iris()
x, y = iris.data, iris.target
```

```python
x
```

    Luego, dividiremos los datos en partes de entrenamiento y prueba.
    15% para probar y 85% para el entrenamiento de algoritmo.

```python
y  # Observamos  que tenemos tres clases 0,1,2
```

```python
x_df = DataFrame(x, columns= iris.feature_names)
```

```python
x_df.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_df, y, test_size=0.15) #test_size=0.15 indica el porcentaje 15 % de datos que se toma para la muestra
```

### Construyendo el modelo
 
Primero, definiremos los parámetros del modelo de regresión como se muestra a continuación. Puede cambiar los valores de acuerdo con sus objetivos de evaluación.

```python
# defining parameters 
params = {
    'boosting': 'gbdt', # Defino el tipo de arbol
    'objective': 'multiclass',  #Defino el objetivo
    'num_leaves': 10,  # Defino el numero de hojas
    'num_class': 3  # Defino el numero de clases
} 
```

A continuación, cargaremos el tren y los datos de prueba en el objeto del conjunto de datos LightGBM. El siguiente código muestra cómo cargar datos de prueba de evaluación y entrenamiento.

```python
# cargando datos
lgb_train = lgb.Dataset(x_train, y_train) # Crea datos de entrenamiento
lgb_eval = lgb.Dataset(x_test, y_test, reference=lgb_train) # Crea datos de validacion
```

Ahora, podemos entrenar el modelo con las variables definidas arriba.

```python
# ajustando el modelo
model = lgb.train(params,
                 train_set=lgb_train,
                 valid_sets=lgb_eval,
                 early_stopping_rounds=30)  # early_stopping_rounds para rondas de parada
```

### Comprobación de predicción y precisión

    Después de entrenar el modelo, podemos predecir los datos de prueba y verificar la precisión de la predicción. Encontraremos el informe de clasificación y la matriz de confusión.

```python
# predicción
y_pred = model.predict(x_test)

y_pred = argmax(y_pred, axis=1)
cr = classification_report(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)

#Imprimo el reporte de confusión
print(cr)
#precisión
#promedio macro
#promedio ponderado 
```

```python
#Imprimo la matriz de confusion
print(cm) 
#La matriz muestra  
# la clase 0 tuvimos todas las prediciones correctas
# La clase 1 tuvimos 6 correctas y 1 falsa
# La clase 2 tuvimos todas las predicciones correctas
```

LightGBM proporciona el método plot_importance() para trazar la importancia de las características. El siguiente código muestra cómo trazarlo.

```python
# Graficamos  la importancia de la característica
lgb.plot_importance(model, height=.5)
```
