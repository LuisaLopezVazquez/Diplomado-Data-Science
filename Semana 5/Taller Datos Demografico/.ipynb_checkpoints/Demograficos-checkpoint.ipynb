# Trabajo con filtros dentro de un DataFrame

```python
#Filtrado sobre las filas
filtro = demograficos.UsuariosInternet < 2 
filtro
```

```python
#Crear filtro2
filtro2 = demograficos.TasaNatalidad > 40
demograficos[filtro2]
```

```python
#para hacerlo en una sola linea
demograficos[demograficos.TasaNatalidad > 40]
```

Queremos saber que paises tienen menos de 2 usuario con acceso a internet y  una tasa de natalidad mayor a 40

```python
#comparar mendiante dos filtros
demograficos[filtro & filtro2]
```

```python
#Aplicar dos filtros en una sola linea
#  &  significa y 
demograficos[  (demograficos.TasaNatalidad > 40) &  (demograficos.UsuariosInternet < 2) ]
```

```python
#obtener el grupo de paises con ingresos altos
demograficos[demograficos.GrupoIngresos == 'Ingreso alto']
```

```python
#Como obtener las categorias
# unique muestra categorias con un unico nombre
demograficos.GrupoIngresos.unique()
```

```python
#Ahora vamos obtener todo acerca de un pais Mexico
demograficos[demograficos.NombrePais == 'Mexico']
```

Acceder a elementos individuales en determinada posición

```python
#Usar funcion iat[]
demograficos.iat[3,4]
```

```python
#Usar funcion iat[0,1] iniciando con la posicion de los renglones, y la siguiente es la columna
demograficos.iat[0,1]
```

```python
#Quiero ver la tasa de natalidad de Angola
demograficos.at[2,'UsuariosInternet']
```

```python
#Por que necesitamos  .at?
filtro3 = demograficos[::10]
filtro3
```

```python
# iat busca en el index en el eje de las y
filtro3.iat[10,0]
```

```python
# at no le interesan los index, busca con el nombre de la columna
filtro3.at[30,'NombrePais']
```

# Iniciando con Matplotlib

```python
#Introducción a Seaborn
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams['figure.figsize'] =  8,4
import warnings
warnings.filterwarnings('ignore')
```

```python
#Mostramos los datos
demograficos.head()
```

```python
#Graficar Numero de usuarios internet
grafica_barras= sns.displot(demograficos['UsuariosInternet'], bins=30)
```

```python
#Vamos a haber los boxplot
#Interpretamos para cada grupo de ingreso donde cae la tasa de natalidad
grafica_bigotes= sns.boxplot(data=demograficos, x='GrupoIngresos', y='TasaNatalidad')
```

#Galeria de seaborn
https://seaborn.pydata.org/examples/index.html

```python
#Crear una grafica de regressión Lineal
grafica_dispersion = sns.lmplot (data=demograficos, x='UsuariosInternet', y='TasaNatalidad', fit_reg=False)
```

Podemos como conclusión en la grafica que entre mayor es la tasa de natalidad 
el numero de usuarios con ecceso a internet diminuye

```python
#Agregamos el grupo de ingresos
grafica_dispersion = sns.lmplot (data=demograficos, x='UsuariosInternet', y='TasaNatalidad', fit_reg=False, hue='GrupoIngresos')
```

Conclusión cuando son paises de ingreso alto tiene mayor acceso a internet, y no tienen tantos hijos
Cuando son paises de ingresos bajos, tienen mas hijos y menos acceso a Internet

```python
#Incrementar el tamaño del grafico mediante size
grafica_dispersion = sns.lmplot (data=demograficos, x='UsuariosInternet', y='TasaNatalidad', fit_reg=False, hue='GrupoIngresos', size=8)
```
