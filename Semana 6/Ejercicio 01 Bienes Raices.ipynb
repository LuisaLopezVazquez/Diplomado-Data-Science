En una curva de campana simétrica, la media, la mediana y la moda tienen el mismo valor. 

```python
#Obtener la media
media= np.mean(df_train['SalePrice'])
media
```

```python
#Obtener la mediana
mediana= np.median(df_train['SalePrice'])
mediana
```

```python
#Importar libreria para obtener la moda
from statistics import mode
```

```python
#Obtener la moda
moda=mode(df_train['SalePrice'])
moda
```

```python
#Imprimimos los valores
print('media:',str(media))
print('mediana:',str(mediana))
print('moda:',str(moda))
```

La media es mayor que la mediana en una distribución positivamente sesgada o asimetria positiva!

La curtosis (o apuntamiento) es una medida de forma que mide cuán escarpada o achatada está una curva o distribución.

## 4.- Confirmando nuestra intuición

### 4.1.- Matriz de Correlación

Lo primero es armar una matriz de correlación (mapa de calor) para ver cuales son las variables más correlacionadas

Vamos declarando una variable corrmat = df_train.corr() con todas las correlaciones entre variables

```python
corrmat = df_train.corr()
f, ax = plt.subplots(1, figsize=(20,9))
sns.heatmap(corrmat, vmax=.8, square=True)
```

La columna que vamos a buscar las variables que se correlacionan es SalePrice
entre mas blanco esten los cuadritos mayor es la correlación por ejemplo: 

*Calidad (OverallQual)

*Area de construccion (GrLivArea)

*Total de autos que caben en el garage (GarageCars)

*Area total del garage(GarageArea)

*Año de construcción(YearBuilt)



```python
#Vamos a tomar las variables que mas se correlacionan con SalePrice
# Matriz de correlación

k = 10  #me muestre 10 columnas
cols = corrmat.nlargest(k, 'SalePrice')['SalePrice'].index
cm= np.corrcoef(df_train[cols].values.T)

sns.set(font_scale=1.15) #Escala de la fuente la letra

hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size':10}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```

Y finalmente hacemos el sns.heatmap

### 4.2.- Matriz de Correlación con Números

Ok, vamos haciendo lo mismo pero ahora con números - y vamos viendo quien se correlaciona más - sin llegar al punto que las 2 variables sean basicamente lo mismo claro

Comenzamos declarando k=10

```python
sns.set(font_scale=1.25)
hm=sns.heatmap(cm, cbar=True, square = True, fmt=".2f",annot_kws={"size":10,}, yticklabels=cols.values,xticklabels=cols.values)
```

Podemos ver que:
- Calidad, AreaVivienda y area Sotano estan muy fuertemente correlacionadas con PrecioVenta

- Sotano y 1er piso son basicamente hermanos gemelos
- Y AñoConstruccion esta ligeramente correlacionada con PrecioVenta, parece que en un futuro habra que hacer un analisis de serie de tiempo más serio.
