# Contexto  y Datos

```python
import pandas as pd
datos = pd.read_csv("ingreso.csv")
datos
```

Variable independiente = horas   
Variable a predecir = ingreso

<BR>
    

```python
import matplotlib.pyplot as plt
plt.ylabel("Ingreso ($)")
plt.xlabel("Promedio de horas semanales trabajadas")
plt.scatter(datos["horas"], datos["ingreso"], color='blue')
plt.show()
```

# Creación del Modelo de
# Regresión Lineal Simple

```python
from sklearn import linear_model

regresion = linear_model.LinearRegression()

horas = datos["horas"].values.reshape((-1, 1)) # -1 termina nuestros elementos 1 es queremos elementos de 1 en 1

#Creamos el modelo
modelo = regresion.fit(horas, datos["ingreso"])

print("Intersección (b)", modelo.intercept_) # intercept_ obtiene la interseccion
print("Pendiente (m)", modelo.coef_) # Obtiene la pendiente

#Vamos a calcular el salario para una persona que trabaja las sig. horas
entrada = [[39.5], [40], [43], [43.5]]
modelo.predict(entrada) #imprimimos y vemos los salarios predecidos

#Graficamos
plt.scatter(entrada, modelo.predict(entrada), color="red")
plt.plot(entrada, modelo.predict(entrada), color="black")

plt.ylabel("Ingreso ($)")
plt.xlabel("Promedio de horas semanales trabajadas")
plt.scatter(datos["horas"], datos["ingreso"], color="blue", alpha=0.55)
plt.show()
```
